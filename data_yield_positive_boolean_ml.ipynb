```
# Importing the dataset
dataset = pd.read_csv('/content/drive/My Drive/My Docs/data_yield_positive_boolean4.csv')
X1 = dataset.iloc[:, 0:9].values
X2 = dataset.iloc[:, 0:9].values
y1 = dataset.iloc[:, 9:10].values
y2 = dataset.iloc[:, 10:11].values
```

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.30, random_state = 0)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size = 0.30, random_state = 0)
```

```
# Part 2 - Now let's make the ANN!

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
# Initialising the ANN
classifier = Sequential()
```

```
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu', input_dim = 9))
# Adding the second hidden layer
classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'linear'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

```

```
classifier.fit( X1_train, y1_train, batch_size = 128, epochs = 12)
classifier.fit( X2_train, y2_train, batch_size = 128, epochs = 12)
```

```
# Predicting the Test set results
y1_pred = classifier.predict(X1_test)
y1_pred = (y1_pred > (y1_pred.sum()/len(y1_pred)))

y2_pred = classifier.predict(X2_test)
y2_pred = (y2_pred > (y2_pred.sum()/len(y2_pred)))
```

```
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y1_test, y1_pred)
print (cm1)
# cm11 = confusion_matrix(y1_train, y1_train_pred)
cm2 = confusion_matrix(y2_test, y2_pred)
print (cm2)

# Error Reporting
from sklearn.metrics import r2_score
y1_error = r2_score(y1_test, y1_pred)
print(y1_error)


from sklearn.metrics import accuracy_score

print (accuracy_score(y1_test, y1_pred))
print (accuracy_score(y2_test, y2_pred))
```
